```python
from selenium import webdriver 
import pandas as pd
import openpyxl

produk = []; URL = []; super_seller = []; pelapak = []; harga = []; diskon = []; stock = []; lokasi = []
feedback_persentase = []; feedback_jumlah = []; img = []; terjual = []; kondisi_barang = []; deskripsi = []
waktu_proses = []; category_url = []; nama_sum = []; tanggal_sum = []; pukul_sum = []
ulasan_sum = []; jumlah_review = []; bintang5 = []; bintang4 = []; bintang3 = []; bintang2 = []; bintang1 = []

#AKSES PAGE
tempat = webdriver.Chrome('E:\chromedriver')
link = 'https://m.bukalapak.com/products/s?from=omnisearch&page=1&search%5Bkeywords%5D=masker+sensi&source=navbar'
tempat.get(link)

#XPATH
produk_ = tempat.find_elements_by_xpath('//*[@id="page"]//article/div[3]/h3/a')
URL_ = tempat.find_elements_by_xpath('//*[@id="page"]//article/div[3]/h3/a[@href]')
pelapak_ = tempat.find_elements_by_xpath('//*[@id="page"]//article/div[3]/div[1]/div/h5/a')
harga_ = tempat.find_elements_by_xpath('//*[@id="page"]//div[3]//div/span[1]/span[2]')
lokasi_ = tempat.find_elements_by_xpath('//*[@id="page"]//article/div[3]//div[@class="ab-product__seller-location product__seller-location u-hidden"]')
feedback_ = tempat.find_elements_by_class_name('user-feedback-summary')
img_ = tempat.find_elements_by_xpath('//*[@id="page"]//article/div[1]/a/picture//img[@src]')
    
#SCRAPING DATA PER PAGE
for x in produk_[:3]:
    produk.append(x.text)
for x in URL_[:3]:
    URL.append(x.get_attribute('href'))
for x in pelapak_[:3]:
    pelapak.append(x.text)
for x in harga_[:3]:
    harga.append(x.text)
for x in lokasi_[:3]:
    lokasi.append(x.text)
for x in feedback_[:3]:
    persentase = int(x.text[:x.text.find('%')])
    feedback_persentase.append(persentase)
    jumlah = int(x.text[x.text.find('(')+1:(x.text.find(' f')+1)])
    feedback_jumlah.append(jumlah)
for x in img_[:3]:
    img.append(x.get_attribute('src'))
    
#SCRAPING DATA PER PRODUK
for y in range(0,len(URL)):
    tempat2 = webdriver.Chrome('E:\chromedriver')
    tempat2.get(URL[y])
    stock_ = tempat2.find_elements_by_xpath('//*[@id="wrapper"]/div[7]/div[3]/div[1]/div[6]/div[2]/div/div[1]/div/div[3]')
    terjual_ = tempat2.find_elements_by_xpath('//*[@id="wrapper"]/div[7]/div[3]/div[1]/div[6]/div[2]/div/div[2]/div/div[3]')
    super_seller_ = tempat2.find_elements_by_xpath('//*[@id="wrapper"]/div[7]/div[3]/div[1]/div[7]/div[2]/div/div[2]/p[3]/span')
    kondisi_barang_ = tempat2.find_elements_by_xpath('//*[@id="wrapper"]/div[7]/div[3]/div[1]/div[6]/div[1]/div/div[2]/div')
    deskripsi_ = tempat2.find_elements_by_class_name('TextEllipsis')
    waktu_proses_ = tempat2.find_elements_by_xpath('//*[@id="wrapper"]/div[7]/div[3]/div[1]/div[7]/div[2]/p[3]')
    category_url_ = tempat2.find_elements_by_xpath('//*[@id="wrapper"]/div[7]/div[3]/div[1]//div/div/ul//li/a/span')
    diskon_ = tempat2.find_elements_by_xpath('//*[@id="wrapper"]/div[7]/div[3]/div[1]/div[4]/div[1]/div[3]/div/p')
    if len(stock_) != 0:
        stock.append(stock_[0].text)
    else:
        stock.append('-')
    if len(terjual_) != 0:
        terjual.append(terjual_[0].text)
    else:
        terjual.append('-')
    if len(kondisi_barang_) != 0:
        kondisi_barang.append(kondisi_barang_[0].text)
    else:
        kondisi_barang.append('-')
    if len(deskripsi_) != 0:
        deskripsi.append(deskripsi_[0].text.replace('\n', ' | '))
    else:
        deskripsi.append('-')
    if len(waktu_proses_) != 0:    
        waktu_proses.append(waktu_proses_[0].text[waktu_proses_[0].text.find('nan ')+4 :])
    else:
        waktu_proses.append('-')
    if len(super_seller_) != 0:
        super_seller.append(1)
    else:    
        super_seller.append(0)
    if len(diskon_) != 0:
        diskon.append(diskon_[0].text.replace('Rp', ''))
    else:
        diskon.append('-')
        
    cu = ''
    for z in category_url_:
        if z.text == produk[y]:
            break
        if z.text == 'Home':
            continue
        else:
            cu = cu + '>' + z.text
    category_url.append(cu) 
        
    #REVIEW
    link_review = '//*[@id="wrapper"]/div[7]/div[3]/div[1]/div[4]/div[1]/div[1]/div[1]/a[@href]'
    webreview_ = tempat2.find_elements_by_xpath(link_review)
    if len(webreview_) != 0: 
        webreview = webreview_[0].get_attribute('href')
        tempat3 = webdriver.Chrome('E:\chromedriver')
        tempat3.get(webreview)
        akun_ = tempat3.find_elements_by_xpath('//*[@id="main"]//article/div[2]')
        ulasan_ = tempat3.find_elements_by_xpath('//*[@id="main"]//article/div[3]')
        jumlah_review_ = tempat3.find_elements_by_xpath('//*[@id="main"]/div/div[1]/div/div/span')
        bintang5_ = tempat3.find_elements_by_xpath('//*[@id="main"]/div/div[2]/div/div[1]/div[3]/span')
        bintang4_ = tempat3.find_elements_by_xpath('//*[@id="main"]/div/div[2]/div/div[2]/div[3]/span')
        bintang3_ = tempat3.find_elements_by_xpath('//*[@id="main"]/div/div[2]/div/div[3]/div[3]/span')
        bintang2_ = tempat3.find_elements_by_xpath('//*[@id="main"]/div/div[2]/div/div[4]/div[3]/span')
        bintang1_ = tempat3.find_elements_by_xpath('//*[@id="main"]/div/div[2]/div/div[5]/div[3]/span')
        
        namaa = ''
        nama = [x.text[x.text.find('h ')+2 : x.text.find(',')] for x in akun_]
        for i in range(len(nama)):
            namaa = namaa + str(i+1) + ') ' + nama[i] + '\n'
        nama_sum.append(namaa)
        
        tanggall = ''
        tanggal = [x.text[x.text.find(', ')+2 : x.text.find(', p')] for x in akun_]
        for i in range(len(tanggal)):
            tanggall = tanggall + str(i+1) + ') ' + tanggal[i] + '\n'
        tanggal_sum.append(tanggall)
        
        pukull = ''
        pukul = [x.text[x.text.find('pukul ')+6 : x.text.find(' WIB')] for x in akun_]
        for i in range(len(pukul)):
            pukull = pukull + str(i+1) + ') ' + pukul[i] + '\n'
        pukul_sum.append(pukull)
        
        ulasann = ''
        for i in range(len(ulasan_)):
            ulasann = ulasann + str(i+1) + ') ' + ulasan_[i].text + '\n'
        ulasan_sum.append(ulasann)

        jumlah_review.append(jumlah_review_[0].text)
        bintang5.append(int(bintang5_[0].text.replace('%', '')))
        bintang4.append(int(bintang4_[0].text.replace('%', '')))
        bintang3.append(int(bintang3_[0].text.replace('%', '')))
        bintang2.append(int(bintang2_[0].text.replace('%', '')))
        bintang1.append(int(bintang1_[0].text.replace('%', '')))

    else:
        nama_sum.append('-')
        tanggal_sum.append('-')
        pukul_sum.append('-')
        ulasan_sum.append('-')
        jumlah_review.append(0)
        bintang5.append(0)
        bintang4.append(0)
        bintang3.append(0)
        bintang2.append(0)
        bintang1.append(0)
        
    tempat2.close()
    tempat3.close()
    
tempat.close()        
print('selesai')
```

```python
#DATAFRAME
masker = {
    'produk' : produk, 
    'category url' : category_url,
    'url' : URL,
    'pelapak' : pelapak,
    'waktu proses' : waktu_proses,
    'harga' : harga,
    'harga setelah diskon' : diskon,
    'lokasi' : lokasi,
    'stock' : stock,
    'terjual' : terjual,
    'kondisi barang' : kondisi_barang,
    'super seller' : super_seller,
    'deskripsi' : deskripsi,
    'feedback (jumlah)' : feedback_jumlah,
    'feedback positif (persen)': feedback_persentase,
    'img' : img,
    'nama review teratas' : nama_sum,
    'tanggal review teratas' : tanggal_sum,
    'pukul review teratas' : pukul_sum,
    'review teratas' : ulasan_sum,
    'jumlah review' : jumlah_review,
    'bintang 5 (persen)' : bintang5,
    'bintang 4 (persen)' : bintang4,
    'bintang 3 (persen)' : bintang3,
    'bintang 2 (persen)' : bintang2,
    'bintang 1 (persen)' : bintang1,
}
dataa = pd.DataFrame(masker)
dataa
```
